```python
import os
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image

main_dir = os.getcwd()
cm_dir = os.path.join(main_dir, 'confusion-matrix-jpeg')
report_dir = os.path.join(main_dir, 'report-jpeg')
cm_expr = os.listdir(cm_dir)
rp_expr = os.listdir(report_dir)
cm_expr_dir = [os.path.join(cm_dir, expr) for expr in cm_expr]
rp_expr_dir = [os.path.join(report_dir, expr) for expr in rp_expr]

cm_files = [os.listdir(expr_dir) for expr_dir in cm_expr_dir]
rp_files = [os.listdir(expr_dir) for expr_dir in rp_expr_dir]

# Create a new Excel workbook
wb = Workbook()

# Check if the 'epoch' sheet exists, and create it if it doesn't
for s_id, sheet_name in enumerate(cm_expr):
    # Check if the sheet already exists
    sheet = wb.create_sheet(sheet_name, s_id + 1)
    
    # Initialize row index for cm_expr
    row_index_cm = 2
    row_index_cm2 = 2

    # Add images from cm_expr to column B
    for i, cm_file in enumerate(cm_files[s_id]):
        # Create a full path to the image file
        image_path = os.path.join(cm_expr_dir[s_id], cm_file)
        if 'test' in cm_file:
            img = Image(image_path)
            sheet.add_image(img, f'B{row_index_cm}')
            img_width, img_height = img.width, img.height
            sheet.column_dimensions['B'].width = img_width / 6  
            sheet.row_dimensions[row_index_cm].height = img_height / 2
        else:
            img = Image(image_path)
            sheet.add_image(img, f'D{row_index_cm2}')
            img_width, img_height = img.width, img.height
            sheet.column_dimensions['D'].width = img_width / 6  
            sheet.row_dimensions[row_index_cm2].height = img_height / 2
            row_index_cm2 += 1

        # Increment the row index for the next image
        row_index_cm += 1

    # Initialize row index for rp_expr
    row_index_rp = 2
    row_index_rp2 = 2

    # Add images from rp_expr to column C
    for i, rp_file in enumerate(rp_files[s_id]):
        # Create a full path to the image file
        image_path = os.path.join(rp_expr_dir[s_id], rp_file)
        if 'test' in rp_file:

        # Add an image to the sheet at the current row index in column C
            img = Image(image_path)
            sheet.add_image(img, f'C{row_index_rp}')
            img_width, img_height = img.width, img.height
            sheet.column_dimensions['C'].width = img_width / 6  
            sheet.row_dimensions[row_index_rp].height = img_height / 1.5

        else:
            img = Image(image_path)
            sheet.add_image(img, f'E{row_index_rp2}')
            img_width, img_height = img.width, img.height
            sheet.column_dimensions['E'].width = img_width / 6  
            sheet.row_dimensions[row_index_rp2].height = img_height / 1.5
            row_index_rp2 += 1

        # Increment the row index for the next image in rp_expr
        row_index_rp += 1

wb.save('experiment.xlsx')
```
